```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# 시드 고정 및 데이터 로드
tf.keras.utils.set_random_seed(42)
tf.config.experimental.enable_op_determinism()

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

# 전처리: 4차원 변환 및 정규화
train_scaled = train_input.reshape(-1, 28, 28, 1) / 255.0
history_model = keras.Sequential([
    # 1. 합성곱 층: 특징 추출
    layers.Conv2D(32, kernel_size=3, activation='relu', padding='same', input_shape=(28, 28, 1)),
    layers.MaxPooling2D(2),
    layers.Dropout(0.25),

    # 2. Flatten: 1차원으로 펼치기
    layers.Flatten(),

    # 3. 은닉층 및 출력층
    layers.Dense(100, activation='relu'),
    layers.Dropout(0.25),  # 과대적합 방지를 위해 추가 권장
    layers.Dense(10, activation='softmax')
])

# 컴파일
history_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 훈련 (검증 데이터 비율 20%)
history = history_model.fit(train_scaled, train_target, epochs=3, validation_split=0.2)

# 시각화: 손실(Loss) 그래프
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```
